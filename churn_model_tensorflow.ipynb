```python
# tensorflow churn prediction model; churn rate = percentage of subscribers of a business that are leaving
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Churn.csv')
```

```python
df.head()
```

```python
# define features and target
X = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1)) # one-hot encoding; convert categorical var into dummy var
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0) # binary classification, 1 = yes 0 = no
```

```python
# split into train and test subsets of 80% and 20%, respectively
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
y_train.head()
```

```python
X_train.head()
```

```python
# import dependencies
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
# build model with two hidden layers
model = Sequential() # instantiate sequential class
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns))) # first hidden layer
model.add(Dense(units=64, activation='relu')) # second hidden layer
model.add(Dense(units=1, activation='sigmoid')) # one output
```

```python
# compile model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
# fit model, predict and evaluate score
X_train=np.asarray(X_train).astype(np.float32)
y_train=np.asarray(y_train).astype(np.float32)
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
# predict
y_hat = model.predict(X_test)
y_hat = [0 if val <0.5 else 1 for val in y_hat] # convert continuous value to binary (0 or 1)
```

```python
y_hat
```

```python
# calculate accuracy score
accuracy_score(y_test, y_hat)
```
